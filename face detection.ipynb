## Face detection

```python
face_cascade = cv2.CascadeClassifier('C:/Users/user/Desktop/DS/opencv/opencv-master/opencv-master/data/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img=img.copy()
    
    face_rect = face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img, (x,y),(x+w,y+h),(255,0,0),4)
    
    return face_img
```

```python
img1 = detect_face(m)
```

```python
plt.imshow(img1, cmap='gray')
```

```python
img2 =detect_face(pilu)
plt.imshow(img2, cmap='gray')
```

```python
img3= detect_face(me)
plt.imshow(img3, cmap='gray')
```

```python
def detect_face2(img):
    
    face_img=img.copy()
    
    face_rect = face_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img, (x,y),(x+w,y+h),(255,0,0),3)
    
    return face_img
```

```python
img4 = detect_face2(pilu)
```

```python
plt.imshow(img4, cmap='gray')
```

## Real time


```python
cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read()
    
    frame = detect_face(frame)
    
    cv2.imshow("Camera", frame)
    
    c= cv2.waitKey(1)
    if c==27:
        break
cap.release()
cv2.destroyAllWindows()
```

## EYE Detection

```python
eye_cascade = cv2.CascadeClassifier('C:/Users/user/Desktop/DS/opencv/opencv-master/opencv-master/data/haarcascades/haarcascade_eye.xml')
```

```python
def detect_eye(img):
    
    face_img=img.copy()
    
    face_rect = eye_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img, (x,y),(x+w,y+h),(255,0,0),5)
    
    return face_img
```

```python
img6=detect_eye(m)
plt.imshow(img6)
```

```python
cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read()
    
    frame = detect_eye(frame)
    
    cv2.imshow("Camera", frame)
    
    c= cv2.waitKey(1)
    if c==27:
        break
cap.release()
cv2.destroyAllWindows()
```
